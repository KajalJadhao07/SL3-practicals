```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("titanic")
```

```python
data.head()
```

```python
sns.distplot(x = data['age'], bins = 10)
```

```python
sns.distplot(data['age'], bins = 10,kde=False)
```

```python
sns.jointplot(x = data['age'], y= data['fare'], kind = 'scatter')
```

```python
sns.jointplot(x = data['age'], y= data['fare'], kind = 'hex')
```

```python
sns.rugplot(data['fare'])
```

```python
data['sex'] = data['sex'].str.capitalize()
sns.barplot(x='sex',y='age', data=data, palette={'Male': 'skyblue', 'Female': 'orange'})
```

```python
sns.countplot(x='sex', data = data,palette={'Male': 'skyblue', 'Female': 'orange'})
```

```python
sns.boxplot(x = 'sex' , y ='age', data = data, palette={'Male': 'skyblue', 'Female': 'orange'})
```

```python
sns.boxplot(x = 'sex' , y ='age', data = data, hue = "survived")
```

```python
sns.violinplot(x='sex', y='age', data = data ,palette={'Male': 'skyblue', 'Female': 'orange'})
```

```python
sns.violinplot(x='sex', y='age' , data = data , hue = "survived")
```

```python
sns.stripplot(x = 'sex' , y='age' , data = data , jitter = True, palette={'Male': 'skyblue', 'Female': 'orange'})
```

```python
sns.stripplot(x = 'sex' , y='age' , data = data, jitter = False , palette={'Male': 'skyblue', 'Female': 'orange'} )
```

```python
sns.stripplot(x = 'sex' , y='age' , data = data, jitter = True , hue = "survived" )
```

```python
sns.swarmplot(x = 'sex', y='age' , data = data , palette={'Male': 'skyblue', 'Female': 'orange'})
```

```python
sns.swarmplot(x='sex', y='age', data = data, hue = 'survived')
```
